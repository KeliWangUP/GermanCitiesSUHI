```python
import os
import glob
import shutil
```

```python
# 1.logic: move all files in the newlstdata folder to the corresponding folder whose suffix is the same as the file prefix
# e.g. move ../../data/newlstdata/67_lst_new.tif to ../../data/processed/Aachen_67 folder and rename it to 67_lst.tif
```

```python
# 2. logic: delete all files in the newcitydata folder named with a suffix as lst.tif 
# and move all the rest to the corresponding folder whose suffix is the same as the file prefix 
# e.g. delete ../../data/newcitydata/67_lst.tif and move ../../data/newcitydata/67_gaia_2018.tif to ../../data/processed/Aachen_67 folder
```

```python
newlstdata_path = '../../data/newlstdata'
processed_path = '../../data/processed'

# Step 1: Scan existing folders in processed_path to build a dynamic ID-to-Folder map
# Expected folder format: "CityName_ID" (e.g., "Aachen_67")
id_to_folder_map = {}

# Iterate through all items in the processed directory
for item in os.listdir(processed_path):
    item_path = os.path.join(processed_path, item)
    
    # Ensure it's a directory and contains an underscore
    if os.path.isdir(item_path) and '_' in item:
        try:
            # Split from the right to handle potential city names with underscores (e.g., "New_York_12")
            prefix_id = item.split('_')[-1]
            id_to_folder_map[prefix_id] = item_path
        except IndexError:
            continue

# Step 2: Iterate through the source files and match them with the mapped directories
search_pattern = os.path.join(newlstdata_path, '*_lst_new.tif')

for file_path in glob.glob(search_pattern):
    filename = os.path.basename(file_path)
    
    # Extract the file prefix ID (e.g., '67' from '67_lst_new.tif')
    file_id = filename.split('_')[0]
    
    # Step 3: Match file ID with the corresponding folder path
    target_city_folder = id_to_folder_map.get(file_id)
    
    if not target_city_folder:
        print(f"Warning: No matching destination folder found for ID '{file_id}' ({filename}). Skipping.")
        continue
        
    # Define the new standardized filename (e.g., '67_lst.tif')
    new_filename = f"{file_id}_lst.tif"
    destination_path = os.path.join(target_city_folder, new_filename)
    
    # Safe move operation
    try:
        shutil.move(file_path, destination_path)
        print(f"Successfully moved: {filename} -> {os.path.join(os.path.basename(target_city_folder), new_filename)}")
    except Exception as e:
        print(f"Error moving file {filename}: {e}")
```

```python
# Define directory paths (assuming id_to_folder_map is already built from step 1)
newcitydata_path = "../../data/newcitydata"

# Ensure the source directory exists before iterating
if os.path.exists(newcitydata_path):
    # Iterate through all files in the newcitydata directory
    for filename in os.listdir(newcitydata_path):
        file_path = os.path.join(newcitydata_path, filename)
        
        # Skip directories, process files only
        if not os.path.isfile(file_path):
            continue
            
        # Action A: Delete files with the suffix '_lst.tif'
        if filename.endswith('_lst.tif'):
            try:
                os.remove(file_path)
                print(f"Successfully deleted: {filename}")
            except Exception as e:
                print(f"Error deleting file {filename}: {e}")
                
        # Action B: Move all other files to their corresponding processed folders
        else:
            # Extract the file prefix ID (e.g., '67' from '67_gaia_2018.tif')
            file_id = filename.split('_')[0]
            
            # Match file ID with the pre-built folder map
            target_city_folder = id_to_folder_map.get(file_id)
            
            if not target_city_folder:
                print(f"Warning: No matching destination folder found for ID '{file_id}' ({filename}). Skipping.")
                continue
                
            destination_path = os.path.join(target_city_folder, filename)
            
            # Safe move operation keeping the original filename
            try:
                shutil.move(file_path, destination_path)
                print(f"Successfully moved: {filename} -> {os.path.basename(target_city_folder)}/")
            except Exception as e:
                print(f"Error moving file {filename}: {e}")
else:
    print(f"Error: Source directory {newcitydata_path} does not exist.")
```
